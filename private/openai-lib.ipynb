```python
def print_markdown(text):
    wrapper = textwrap.TextWrapper(width=80)  # Set the wrap width to 80 characters
    wrapped_text = wrapper.fill(text)
    
    # Markdown formatting
    markdown_text = wrapped_text.replace("* ", "- **") + "**"  # Convert bullet points to bold Markdown list items
    
    # Print wrapped and formatted text
    print(markdown_text)
```
